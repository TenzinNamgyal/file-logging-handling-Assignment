Q15:- How does Python's garbage collection system work?

Ans 15:- Python’s garbage collection system works by automatically freeing memory of objects that are no longer in use using reference counting and generational garbage collection to handle circular references and improve memory efficiency.  
#### 1. Reference Counting

    Each object in Python keeps a count of references pointing to it.

    When the reference count drops to zero, the object is no longer accessible and is immediately deallocated.  
#### 2. Handling Circular References

    Sometimes objects reference each other, forming cycles.

    Reference counting alone cannot free these objects.

    Python’s garbage collector detects and removes circular references.  
#### 3. Generational Garbage Collection

    Python uses generational GC: objects are grouped by age (young, middle-aged, old)

    Younger objects are collected more frequently because they tend to become unreachable faster.

    Older objects are checked less often to save time.  
#### 4. Garbage Collection Modules

    gc module allows you to:

    Inspect objects

    Enable/disable GC

    Force collection

Q23:- What are the challenges associated with memory management in Python?

Ans23:- 
### Key Challenges
#### 1. Circular References

    Objects referencing each other can prevent memory from being freed

    Reference counting alone cannot handle this

    Garbage collector solves this, but may not always run immediately  
    
#### 2. Memory Leaks

    Occur when unused objects are still referenced

    Can slowly consume memory in long-running programs

#### 3. High Memory Usage

    Python stores objects as full objects in memory

    For example, lists, dictionaries, and strings can use more memory than equivalent low-level structures in C

#### 4. Garbage Collection Overhead

    Garbage collection runs periodically

    Can cause performance overhead, especially in programs with many objects

#### 5. Large Data Structures

    Handling large datasets (big lists, arrays) may consume significant memory

    Need efficient data structures or external storage

#### 6. Manual Management of External Resources

    Python only manages Python objects

    Files, sockets, and database connections still need explicit closure (with statement or close())

Q18:- How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)  # Log INFO and above

# Create a rotating file handler
handler = RotatingFileHandler(
    "app.log",      # Log file name
    maxBytes=1_000_000,  # Rotate after 1MB
    backupCount=3        # Keep up to 3 old log files
)

# Create a formatter and add it to the handler
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example log messages
logger.info("Program started")
logger.warning("This is a warning")
logger.error("This is an error")
```

Q21:- Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
# File to read
filename = "sample.txt"
# Word to search for
word_to_count = "Python"

try:
    with open(filename, "r") as file:
        content = file.read()  # Read entire file content

    # Count occurrences (case-sensitive)
    count = content.count(word_to_count)

    print(f"The word '{word_to_count}' occurs {count} time(s) in the file.")

except FileNotFoundError:
    print(f"Error: The file '{filename}' does not exist.")
```

Q22:-How can you check if a file is empty before attempting to read its contents?

```python
import os

filename = "empty.txt"

if os.path.exists(filename) and os.stat(filename).st_size > 0:
    with open(filename, "r") as file:
        content = file.read()
        print("File contents:")
        print(content)
else:
    print("The file is empty")
```

Q23:- Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging

# Configure logging to write to a file
logging.basicConfig(
    filename="file_errors.log",   # Log file name
    level=logging.ERROR,          # Log only errors
    format="%(asctime)s - %(levelname)s - %(message)s"
)

filename = "nonexistent_file.txt"

try:
    # Attempt to open a file for reading
    with open(filename, "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError as e:
    print(f"Error: The file '{filename}' was not found.")
    logging.error(f"FileNotFoundError: {e}")  # Log the error

except PermissionError as e:
    print(f"Error: Permission denied for file '{filename}'.")
    logging.error(f"PermissionError: {e}")  # Log the error

except Exception as e:
    print(f"An unexpected error occurred: {e}")
    logging.error(f"Unexpected error: {e}")  # Log the error
```
